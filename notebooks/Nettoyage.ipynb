## Chargement dataset

```python
import pandas as pd

dfEco=pd.read_csv('../Data Collection/economic_world_cup_data.csv')
dfEnv=pd.read_csv('../Data Collection/environmental_world_cup_data.csv')
dfSoc=pd.read_csv('../Data Collection/social_indicators.csv')
dfUrb=pd.read_csv('../Data Collection/urban_infrastructure_data.csv')
dfWhr=pd.read_csv('../Data Collection/WHR.csv',sep=';')
dfFifa=pd.read_csv('../Data Collection/worldCupData.csv')
```

## Filtrage et harmonisation des années par pays
- Germany => [2003;2009]
- South Africa => [2007;2013]
- Brazil => [2011;2017]
- Rusia => [2015;2021]
- Qatar => [2019;2025]

### Data ECO

```python
#  Germany => [2003;2009]

germny_data = dfEco[dfEco['Country'] == 'DEU']
germny_data = germny_data[(germny_data['Year'] >= 2003) & (germny_data['Year'] <= 2009)]
germny_data
```

```python
# South Africa => [2007;2013]

ZAF_data = dfEco[dfEco['Country'] == 'ZAF']
ZAF_data = ZAF_data[(ZAF_data['Year'] >= 2007) & (ZAF_data['Year'] <= 2013)]
ZAF_data
```

```python
# Brazil => [2011;2017]
BRA_data = dfEco[dfEco['Country'] == 'BRA']
BRA_data = BRA_data[(BRA_data['Year'] >= 2011) & (BRA_data['Year'] <= 2017)]
BRA_data
```

```python
# Russia => [2015;2021]

RUS_data = dfEco[dfEco['Country'] == 'RUS']
RUS_data = RUS_data[(RUS_data['Year'] >= 2015) & (RUS_data['Year'] <= 2021)]
RUS_data
```

```python
# Qatar => [2019;2025]

QAT_data = dfEco[dfEco['Country'] == 'QAT']
QAT_data = QAT_data[(QAT_data['Year'] >= 2019) & (QAT_data['Year'] <= 2025)]
QAT_data
```

```python
# Concaténation des données économiques pour les 5 pays hôtes
dfEco = pd.concat([germny_data, ZAF_data, BRA_data, RUS_data, QAT_data], ignore_index=True)
dfEco.value_counts('Country')
```

```python
print("les valeurs nulls dans le dataset économique :",dfEco.isnull().sum())
print("\n\nles doublants dans le dataset économique :",dfEco.duplicated().sum())
```

```python
dfEco.to_csv('../Data Clean/cleaned_economic_world_cup_data.csv', index=False)
```
